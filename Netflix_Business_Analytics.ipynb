```python
#I am going to impute values to the missing values in the dataset to maintain the consistency in the data set.
df['director'] = df['director'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna('Unknown')
```

```python
df = df.dropna(subset=['rating', 'duration'])
```

```python
#check to missing percentage in the dataset
missing_percent = df.isnull().mean() * 100
print(missing_percent)
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors='coerce')
```

```python
df["date_added"].dtype
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['day_added'] = df['date_added'].dt.day
```

```python
df_2020 = df[df['year_added'] == 2020]
```

```python
df.groupby(df['year_added']).size().plot(kind='bar')
```

```python
#create a release decade so that i can manage to derive the trend accordingly to the decade
df["Release_Decade"] = (df["release_year"] // 10) * 10
```

```python
df["Movie_Duration"] = np.where(
    df["type"] == "Movie",
    df["duration"].str.replace(" min", "", regex=False),
    np.nan
)
```

```python
df["Movie_Duration"] = pd.to_numeric(df["Movie_Duration"])
```

```python
df[["duration","Movie_Duration"]].head()
```

```python
# one title may belong to the nultiple genre so here i split the genre so that i can count them .
df["Genre_Count"] = df["listed_in"].str.split(",").str.len()
```

```python
# creating the cast_count to count number of the cast being part of that title.
df["Cast_Count"] = df["cast"].str.split(",").str.len()
```

```python
df["Country_Count"] = df["country"].str.split(",").str.len()
```

```python
df["Director_Count"] = df["director"].str.split(",").str.len()
```

```python
# Content age to identify the time that how old the content was?
current_year = pd.Timestamp.now().year

df["Content_Age"] = current_year - df["release_year"]
```

```python
import datetime

current_year = datetime.datetime.now().year

# Make sure the column exists
df['year_added'] = df['date_added'].dt.year

# Calculate how many years each title has been on Netflix
df['Years_On_Netflix'] = current_year - df['year_added']
```

```python
df["Content_Category"] = np.where(
    df["Content_Age"] <= 5,
    "New",
    "Old"
)
```

```python
# what is the overall composition of Netflix's content library?
total_titles = len(df)
total_movies = (df["type"] == "Movie").sum()
total_tv = (df["type"] == "TV Show").sum()
total_countries = df["country"].nunique()
total_directors = df["director"].nunique()
total_genres = df["listed_in"].nunique()

print(f"Total Titles: {total_titles}")
print(f"Movies: {total_movies}")
print(f"TV Shows: {total_tv}")
print(f"Countries: {total_countries}")
print(f"Directors: {total_directors}")
print(f"Genres: {total_genres}")
```

```python
# Make sure the column exists
df['year_added'] = df['date_added'].dt.year

# Group by year_added instead of Added_Year
growth = (
    df.groupby("year_added")
      .size()
      .reset_index(name="Titles")
)

print(growth.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(growth['year_added'], growth['Titles'], color='skyblue')
plt.title("Netflix Catalog Growth Over Time")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.show()
```

```python
growth["Growth %"] = growth["Titles"].pct_change()*100
growth
```

```python
# Which countries focus on Movies and which focus on TV Shows?
country_type = (
    df.assign(country=df["country"].str.split(", "))
      .explode("country")
      .pivot_table(
          index="country",
          columns="type",
          values="show_id",
          aggfunc="count",
          fill_value=0
      )
)

country_type.head()
```

```python
country_type.sort_values("Movie",ascending=False).head(15)
```

```python
# Which countries produce longer movies?
movies = df[df["type"]=="Movie"].copy()

movies["Movie_Duration"] = (
    movies["duration"]
    .str.replace(" min","", regex=False)
    .astype(int)
)

country_duration = (
    movies.assign(country=movies["country"].str.split(", "))
          .explode("country")
          .groupby("country")["Movie_Duration"]
          .mean()
          .sort_values(ascending=False)
)

country_duration.head(20)
```

```python
# Top Directors by Average Movie Duration
director_duration = (
    movies.groupby("director")["Movie_Duration"]
          .mean()
          .sort_values(ascending=False)
)

director_duration.head(15)
```
